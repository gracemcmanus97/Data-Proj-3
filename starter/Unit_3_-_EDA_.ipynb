#### Fill in blanks with a brief description of each plot type:

  
- Histogram
> Similar to a bar chart. A frequency distribution chart that shows how often each different value in a set of data occurs.  
- Density Plot / KDE
> Continuous and smoothed version of the histogram. Visualizes the distribution of data over a given period, peaks showing where values are concentrated. 
- Bar Chart
> Vertical or horizontal bars on a graph where the length of the bars are proportional to the measure of data. 
- Line Plot
> Displays information as a series of data points called 'markers' connected by straight line segments. 
- Univariate scatter plot
> Displays the values of a single variable as symbols plotted along a line. Allows you to see the spread of the data. 
- Violin Plot
> Hybrid of a box plot and a kernel density plot, which shows peaks in data. Used to visualize distribution of numerical data. Can show statistics but the density of each variable. 
- Box Plot
> Graphical rendition of a statistical data bsed on the minimum, first quartile, median, third quartile, and maximum. 
- Pie Chart
> Type of graph in which a circle is divided into sectors that each represent a proportion of the whole. 
- Rug Plot
> Can be used to add to your primary plot. Simply a projection of raw data points onto a particular axis, without involving the binning process of a histogram.

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

```python
# import pearson and spearmanr methods from scipy.stats library
from scipy.stats import pearsonr, spearmanr
```

### What is the correlation between salary and education?

#### Your Hypothesis = My hypothesis is the more education you have, the higher the salary you will receive. 

```python
# calculate correlation and pvalue of salary and education with spearmanr method
spearmanr(df['salary'], df['education'])
```

There is a positive correlation between education received and salary.

#### Are these results statistically significant? How do you know?

The correlation is statistically significant because the pvalue is 0. Anything lower than .05 is considered significant. 

### What is the correlation between salary and age?

#### Your Hypothesis: My hypothesis is the older you get the higher the salary received will be. 

```python
# calculate correlation and pvalue of salary and age with pearsonr method
pearsonr(df['salary'], df['age'])
```

Positive correlation

#### Are these results statistically significant? How do you know? 

Yes because it is over .05

### What is the correlation between salary and industry experience? 

#### Your Hypothesis = The more industry experience you have the higher salary received will be. 

```python
# calculate correlation and pvalue of salary and industry experience with pearsonr method
spearmanr(df['salary'], df['industry_experience'])
```

Positive correlation

#### Are these results statistically significant? How do you know?

Yes because it is over .05

#### Crosstab of average salary grouped by education and industry

```python
# use pd.crosstab() to calculate mean salary in each industry grouped by education level
pd.crosstab(index=df['industry'], columns=df['education'], values=df['salary'], aggfunc=np.mean).round(0)
```

#### Heatmap of crosstab of average salary grouped by education and industry

```python
# save above result to new variable
crosstab = pd.crosstab(index=df['industry'], columns=df['education'], values=df['salary'], aggfunc=np.mean).round(0)
```

```python
# create heatmap with annotated mean salary in each industry and education level
plt.figure(figsize=(16,10))
ax = sns.heatmap(crosstab, annot=True, linewidths=.5, cmap='YlGnBu', fmt='g', xticklabels=['High School', 'Some College', "Bachelor's Degree", "Master's Degree"])
plt.xlabel("Education Level", fontsize=14)
plt.ylabel("Industry", fontsize=14)
plt.title("Manager Salaries Across Industries by Education Level", fontsize=18);
```

##### Insights:

If an individual has a masters degree, then they will have the highest paid salary.
